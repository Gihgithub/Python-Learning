```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Apr 28 15:46:31 2019

@author: berkunis
"""
##############################################01_02_PythonLibraries#####################################################
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt 

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder


from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor


from sklearn.metrics import r2_score

from sklearn.svm import SVR
#import data
#df_insurance = pd.read_csv("../Datasets/insurance.csv")
df_insurance = pd.read_csv("Datasets/insurance.csv")

#see the first 15 lines of data
print(df_insurance.head(15))
```

## NULL Values

```python
df_insurance.info()
```

```python
df_insurance.isnull().sum()
```

```python
df_insurance = df_insurance.dropna()
df_insurance.info()
```

## Dummy Encording

```python
# Dummy Encoding Region
df_insurance.region.unique()
encoded_data_r = pd.get_dummies(df_insurance['region'])
encoded_data_r
```

```python
# Dummy Encoding sex
df_insurance.region.unique()
encoded_data_s = pd.get_dummies(df_insurance['sex'])
encoded_data_s
```

```python
# Dummy Encoding smoker
df_insurance.region.unique()
encoded_data_smo = pd.get_dummies(df_insurance['smoker'])
encoded_data_smo
```

## Putting the data together:

```python
##take the numerical data from the original data
X_num = df_insurance[['age', 'bmi', 'children']].copy()

##take the encoded data and add to numerical data
X_final = pd.concat([X_num, encoded_data_r, encoded_data_s, encoded_data_smo ], axis = 1)

#define y as being the "charges column" from the original dataset
y_final = df_insurance[['charges']].copy()

#Test train split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.33, random_state = 0 )
```

## FEATURE SCALLING
### Dialute the bias comming from the outliers


```python

# normalized scaler (fit transform on train, fit only on test)
#n_scaler = MinMaxScaler()
#X_train = n_scaler.fit_transform(X_train.astype(np.float))
#X_test= n_scaler.transform(X_test.astype(np.float))


# standard scaler (fit transform on train, fit only on test)
s_scaler = StandardScaler()
X_train = s_scaler.fit_transform(X_train.astype(np.float64))
X_test= s_scaler.transform(X_test.astype(np.float64))
X_test
```

## Regressional prediction

### Linear Regression

```python
# Appling Linear Regression
# Linear relationshsip


lr = LinearRegression().fit(X_train,y_train)
y_train_predict = lr.predict(X_train)
y_test_predict = lr.predict(X_test)

#print score
print("lr.coef_: {}".format(lr.coef_))
print("lr.intercept_: {}".format(lr.intercept_))
print('lr train score %.3f, lr test score: %.3f' % (
lr.score(X_train,y_train),
lr.score(X_test, y_test)))
```

### Polynomial Regression

```python
# Non linear relationship 

poly = PolynomialFeatures (degree = 2)
X_poly = poly.fit_transform(X_final)

X_train,X_test,y_train,y_test = train_test_split(X_poly,y_final, test_size = 0.33, random_state = 0)

#standard scaler (fit transform on train, fit only on test)
sc = StandardScaler()
X_train = sc.fit_transform(X_train.astype(np.float64))
X_test= sc.transform(X_test.astype(np.float64))

#fit model
poly_lr = LinearRegression().fit(X_train,y_train)

y_train_pred = poly_lr.predict(X_train)
y_test_pred = poly_lr.predict(X_test)

#print score
print('poly train score %.3f, poly test score: %.3f' % (
poly_lr.score(X_train,y_train),
poly_lr.score(X_test, y_test)))
```

## Decision Tree 

```python
# Simlistic nature. Prone to overfitting and low scores in test data


dt = DecisionTreeRegressor(random_state=0)

#test train split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.33, random_state = 0 )

#standard scaler (fit transform on train, fit only on test)
sc = StandardScaler()
X_train = sc.fit_transform(X_train.astype(np.float64))
X_test= sc.transform(X_test.astype(np.float64))


#fit model
dt = dt.fit(X_train,y_train.values.ravel())
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)

#print score
print('dt train score %.3f, dt test score: %.3f' % (
dt.score(X_train,y_train),
dt.score(X_test, y_test)))
```

## Randon Forest Regression

##### Random forest segments the data and applies decision tree algorithm to all data segments and agreegate the results. Hence Random forret is better than Decision tree, does not need data scalling(Normalization)

```python
forest = RandomForestRegressor(n_estimators = 100,
                              criterion = 'mse',
                              random_state = 1,
                              n_jobs = -1)
#test train split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.33, random_state = 0 )

#standard scaler (fit transform on train, fit only on test)
sc = StandardScaler()
X_train = sc.fit_transform(X_train.astype(np.float64))
X_test= sc.transform(X_test.astype(np.float64))

#fit model
forest.fit(X_train,y_train.values.ravel())
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)

#print score
print('forest train score %.3f, forest test score: %.3f' % (
forest.score(X_train,y_train),
forest.score(X_test, y_test)))
```

## Hyperparameter Optimization

### Tring to find the Optimum algorithm parameters to increase the prediction score. 

```python
###Challenge 2:Decision Tree parameter grid###
param_grid_dt = dict(min_samples_leaf=np.arange(9, 13, 1, int), 
                  max_depth = np.arange(4,7,1, int),
                  min_impurity_decrease = [0, 1, 2],
                    )

dt = GridSearchCV(DecisionTreeRegressor(random_state=0), param_grid=param_grid_dt, cv=5,  verbose=3)



#fit model
dt = dt.fit(X_train,y_train.values.ravel())


#print score
print('\n\ndt train score %.3f, dt test score: %.3f' % (
dt.score(X_train,y_train),
dt.score(X_test, y_test)))

print(dt.best_params_)


```

```python


###Challenge 3:Random Forest parameter grid###
param_grid_rf = dict(n_estimators=[20],
                     max_depth=np.arange(1, 13, 2),
                     min_samples_split=[2],
                     min_samples_leaf= np.arange(1, 15, 2, int),
                     bootstrap=[True, False],
                     oob_score=[False, ])


forest = GridSearchCV(RandomForestRegressor(random_state=0), param_grid=param_grid_rf, cv=5, verbose=3)

#fit model
forest.fit(X_train,y_train.values.ravel())


#print score
print('\n\nforest train score %.3f, forest test score: %.3f' % (
forest.score(X_train,y_train),
forest.score(X_test, y_test)))


print(forest.best_params_)
```
