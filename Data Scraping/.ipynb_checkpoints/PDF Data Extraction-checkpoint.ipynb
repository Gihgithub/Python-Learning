##  Extracting Data from a PDF File 

```python
# conda update -n base -c defaults conda
```

```python
import PyPDF2 as p2
```

```python
# creating a pdf file object
PDFfile = open("TM Toolbox Oct2021.pdf","rb")
# creating a pdf reader object
pdfread = p2.PdfFileReader(PDFfile)

# Extract Single Page & print
x = pdfread.getPage(0)
print(x.extractText())
```

```python
# Get details of the PDF file
print(pdfread.getIsEncrypted())
print(pdfread.getDocumentInfo())
print(pdfread.getNumPages())
```

```python
# Extract the entire pdf file
i = 0
while i<pdfread.getNumPages():
        pageinfo = pdfread.getPage(i)
        print(pageinfo.extractText())
        i = i + 1
```
